FOOTBALL PLAYER MANAGEMENT SYSTEM

```python
class Player:
    def __init__(self, pid, pname, ppos, pteam):
        self.id = pid
        self.name = pname
        self.position = ppos
        self.team = pteam

    def add(self):
        try:
            with open("players.txt", "a") as f:
                f.write(self.id + "\t")
                f.write(self.name + "\t")
                f.write(self.position + "\t")
                f.write(self.team + "\n")
                print("Player information added.")
        except IOError:
            print("Error: Could not write to file.")

    def show(self):
        try:
            with open("players.txt", "r") as f:
                print("ID\tName\tPosition\tTeam")
                print("------------------------------------")
                print(f.read())
        except IOError:
            print("Error: Could not read file.")

    def search(self, search_param):
        try:
            with open("players.txt", "r") as f:
                all = f.readlines()
                found = False
                for data in all:
                    if search_param.lower() in data.lower():
                        print(data)
                        found = True
                if not found:
                    print("No player information found matching the search parameter.")
        except IOError:
            print("Error: Could not read file.")

    def delete(self, i):
        try:
            with open("players.txt", "r") as f:
                all = f.readlines()
            with open("players.txt", "w") as f:
                for data in all:
                    d = data.split("\t")
                    if d[0] != i:
                        f.write(data)
                print("Player information deleted.")
        except IOError:
            print("Error: Could not read or write file.")

    def update(self, i):
        try:
            with open("players.txt", "r") as f:
                all = f.readlines()
            with open("players.txt", "w") as f:
                for data in all:
                    d = data.split("\t")
                    if d[0] == i:
                        nn = input("Enter new name: ")
                        np = input("Enter new position: ")
                        nt = input("Enter new team: ")
                        f.write(d[0] + "\t" + nn + "\t" + np + "\t" + nt + "\n")
                    else:
                        f.write(data)
                print("Player information updated.")
        except IOError:
            print("Error: Could not read or write file.")

while True:
    print("Welcome to the Football Player Management System.")
    print("1. Add player information")
    print("2. Delete player information")
    print("3. Update player information")
    print("4. Search player information")
    print("5. Show all player information")
    print("6. Exit")
    try:
        ch = int(input("Enter your choice: "))
        if ch == 1:
            pid = input("Enter player ID: ")
            pname = input("Enter player name: ")
            ppos = input("Enter player position: ")
            pteam = input("Enter player team: ")
            p = Player(pid, pname, ppos, pteam)
            p.add()
        elif ch == 2:
            i = input("Enter player ID to delete: ")
            p = Player("", "", "", "")
            p.delete(i)
        elif ch == 3:
            i = input("Enter player ID to update: ")
            p = Player("", "", "", "")
            p.update(i)
        elif ch == 4:
            search_param = input("Enter search parameter: ")
            p = Player("", "", "", "")
            p.search(search_param)
        elif ch == 5:
            p = Player("", "", "", "")
            p.show()
        elif ch == 6:
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")
    except Exception as e:
        print("An error occurred:", e)
```
